```python
## creating independent feacture
X=data.drop("quality",axis=1)
```

```python
X.head()
```

```python
## creating dependent feature
y=data.quality
```

```python
y.head()
```

SVC model training without Hyper-parameter Tuning

```python
## Splitting the dataset into training set and test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.25,random_state=42)
```

```python
## checking top 5 rows
X_train.head()
```

```python
## printing counts
print(f"X_train is {len(X_train)}, y_train is {len(y_train)}")
print(f"X_test is {len(X_test)}, y_test is {len(y_test)}")
```

```python
##standardization of data
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
## scaling of train data
X_train =scaler.fit_transform(X_train)
X_train
```

```python
y
```

```python
##scaling of test data
X_test =scaler.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression
```

```python
regression.fit(X_train,y_train)


## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## Prediction for the test data
reg_pred=regression.predict(X_test)
reg_pred
```

R square and adjusted R square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

Accuracy of wine quality prediction using Linear Regression is 35 percent.

Support Vector Regressor

```python
## Fitting kernel SVM to the Training set
from sklearn.svm import SVR
model=SVR(kernel='rbf')
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
## Predicting test results
y_predict= model.predict(X_test)
y_predict
```

```python
# importing mean_squared_error, mean_absolute_error

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

##  printing Mean Squared Error
print("Mean Square Error:",mean_squared_error(y_test, y_predict))

##  printing Mean Absolute Error
print("Mean Absolute Error:",mean_absolute_error(y_test, y_predict))

##  printing Root Mean Squared Error
print("Root Mean Squared Error:",np.sqrt(mean_squared_error(y_test, y_predict)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, r2_score
svr_r2_score = r2_score(y_test,y_predict)
print ("Model has {}% accuracy".format((svr_r2_score)*100))
```

```python
##  Adjusted R Square
adjusted_r2=1-(1-r2_score(y_test,y_predict))*(len(y)-1)/(len(y)-X.shape[1]-1)
print ("Model's adjusted_r2 score is:",adjusted_r2)
## R sqaure is adjusted by the number of independent variables.
```

```python
## applying Grid Search to find the best model(better accuracy)
from sklearn.model_selection import GridSearchCV

## defining parameter range
parameters ={'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
grid_search= GridSearchCV(SVR(), parameters ,refit =True, verbose=3)

## fitting the model for grid search
grid_search = grid_search.fit(X_train, y_train)
grid_search
```

```python
## printing best parameter after tuning
print(grid_search.best_params_)
```

```python
## model after hyper-tuning
print(grid_search.best_estimator_)
```

```python
## predictions
y_predict2=grid_search.predict(X_test)
y_predict2
```

```python
##  printing Mean Squared Error
print("Mean Square Error:",mean_squared_error(y_test, y_predict2))

##  printing Mean Absolute Error
print("Mean Absolute Error:",mean_absolute_error(y_test, y_predict2))

##  printing Root Mean Squared Error
print("Root Mean Squared Error:",np.sqrt(mean_squared_error(y_test, y_predict2)))
```

```python
svr_r2_score = r2_score(y_test,y_predict2)
print ("Model has {}% accuracy".format((svr_r2_score)*100))
```

```python
##  Adjusted R Square
adjusted_r2=1-(1-r2_score(y_test,y_predict2))*(len(y)-1)/(len(y)-X.shape[1]-1)
print ("Model's adjusted_r2 score is:",adjusted_r2)
## R sqaure is adjusted by the number of independent variables.
```
